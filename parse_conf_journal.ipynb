```python
import requests

def get_journal_info(doi):
    headers = {
        "Accept": "application/vnd.citationstyles.csl+json"
    }
    response = requests.get(f"https://doi.org/{doi}", headers=headers)
    
    if response.status_code == 200:
        data = response.json()
        return {
            "title": data.get("title"),
            "journal": data.get("container-title"),
            "issn": data.get("ISSN"),
            "publisher": data.get("publisher"),
            "published_date": data.get("issued")
        }
    else:
        return "DOI not found or error in retrieval"

# Example usage
doi = "10.1000/exampledoi"
doi = "10.1016/j.neunet.2021.05.001"
doi = "10.1109/cvpr.2017.602"
doi = "10.22215/etd/2020-14050"
doi = "10.1109/cvpr.2017.702"
doi = "10.7554/elife.29605.005"
doi = "10.1007/978-3-319-43455-1_7"
info = get_journal_info(doi)
pprint(info)
```

```python
import requests

def get_publication_info(doi):
    api_url = f"https://api.crossref.org/works/{doi}"
    response = requests.get(api_url)
    
    if response.status_code == 200:
        data = response.json()
        work = data['message']
        # The 'event' field might contain the conference information if available
        event = work.get('event', {})
        # 'container-title' usually holds the journal or proceedings title
        container_title = work.get('container-title', [])
        
        return {
            "title": work.get("title"),
            "event": event.get("name"),
            "conference_acronym": event.get("acronym"),
            "conference_theme": event.get("theme"),
            "journal": container_title[0] if container_title else None,
            "issn": work.get("ISSN"),
            "publisher": work.get("publisher"),
            "published_date": work.get("issued")
        }
    else:
        return "DOI not found or error in retrieval"
'''
# Example usage
doi = "10.1109/CVPR.2001.990517"

doi = "10.1000/exampledoi"
doi = "10.1016/j.neunet.2021.05.001"
doi = "10.1109/cvpr.2017.602"
doi = "10.1109/cvpr.2017.702"
doi = "10.1007/978-3-319-43455-1_7"
doi = "10.7554/elife.29605.005"
doi = "10.22215/etd/2020-14050"
'''

for k, v in tqdm(whole_paper_dict.items()) :
    if v.DOI is None :
        continue
    info = get_publication_info(v.DOI)
    if isinstance(info, str) :
        continue
    v.conference_acronym = info["conference_acronym"]
    v.publisher = info["publisher"]
```
